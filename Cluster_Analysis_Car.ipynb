```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
import scipy.cluster.hierarchy as shc


# Importing the CSV File
search_data = pd.read_csv('D:\Desktop\simulation\Car_Curated.csv')

# Removing Percent symbols from the columns
search_data['Toyota'] = list(map(lambda x: x[:-1], search_data['Toyota'].values))
search_data['Toyota'] = [float(x) for x in search_data['Toyota'].values]
search_data['Volkswagen'] = list(map(lambda x: x[:-1], search_data['Volkswagen'].values))
search_data['Volkswagen'] = [float(x) for x in search_data['Volkswagen'].values]
search_data['Ford'] = list(map(lambda x: x[:-1], search_data['Ford'].values))
search_data['Ford'] = [float(x) for x in search_data['Ford'].values]
search_data['Honda'] = list(map(lambda x: x[:-1], search_data['Honda'].values))
search_data['Honda'] = [float(x) for x in search_data['Honda'].values]
search_data['Kia'] = list(map(lambda x: x[:-1], search_data['Kia'].values))
search_data['Kia'] = [float(x) for x in search_data['Kia'].values]

# Converting all the data to an Array
data = search_data.iloc[:,1:].values

# Extracting all the Region names
labels = search_data.loc[:,'Country'].values

# Plotting the Dendrogram
plt.figure(figsize=(15, 10))
plt.title("Search Dendrogram")
plt.xlabel('Countries')
plt.ylabel('Euclidean Distance')
dend = shc.dendrogram(shc.linkage(data, method='ward'), labels=labels, orientation='top', leaf_font_size=8)
```
